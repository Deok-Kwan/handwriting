```python
data = pd.read_csv('/data/csafeproject/another_datas/CSAFE/naver_ocr.csv')
data['image_path'] = data['image_path'].str.replace('/content/xai_train', '/data/csafeproject/another_datas/CSAFE/csafe_version5_xai_train', regex=False)
data['repeat'] = data['image_path'].apply(lambda x: int(re.search(r'_(\d+)\.png', x).group(1)) if re.search(r'_(\d+)\.png', x) else None)
data = data[data['label'] > 299].reset_index(drop=True)
data
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
from PIL import Image
import pandas as pd
from tqdm import tqdm  # tqdm을 불러옵니다.
# Custom Dataset 정의
class addfeature(Dataset):
    def __init__(self, df, transform=None):
        self.df = df
        self.transform = transform

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        image_path = self.df.iloc[idx]['image_path']
        image = Image.open(image_path).convert('RGB')
        if self.transform:
            image = self.transform(image)
        return image, idx
# Dataset과 DataLoader 정의
dataset = addfeature(data, transform=transform)
dataloader = DataLoader(dataset, batch_size=64, shuffle=False)

# 모델을 이용한 벡터 추출
vector_list = []

with torch.no_grad():
    for images, indices in tqdm(dataloader, desc="Processing batches"):  # tqdm을 적용하여 진행 상황을 표시
        images = images.to(device)
        outputs = modified_model.get_latent_vector(images)  # 70차원 벡터 추출
        outputs = outputs.cpu().numpy()  # CPU로 이동 후 numpy로 변환
        for i, idx in enumerate(indices):
            vector_list.append((idx, outputs[i]))

# 데이터프레임에 벡터 추가
vector_list = sorted(vector_list, key=lambda x: x[0])  # 인덱스 순으로 정렬
vector_data = [x[1] for x in vector_list]
vector_columns = [f'var_{i+1}' for i in range(128)]
df_vectors = pd.DataFrame(vector_data, columns=vector_columns)

# 원본 데이터프레임에 합치기
data = pd.concat([data, df_vectors], axis=1)
# 원하는 열 순서
column_order = ['image_path', 'label', 'repeat', 'text']
# 나머지 열 자동 추가
column_order += [col for col in data.columns if col not in column_order]
# 열 순서 변경
data = data[column_order]
data = data.sort_values(by=['label', 'repeat','text']).reset_index(drop=True)

data
```
